# 1. Build graph using TF operations

```python
import tensorflow as tf
```

```python
# X and Y data
x_train = [1, 2, 3]
y_train = [1, 2, 3]
```

```python
W = tf.Variable(tf.random_normal([1]), name='weight')  # [1] : 값이 하나인 1차원 array를 나타내는 shape 명시
b = tf.Variable(tf.random_normal([1]), name='bias')
```

> tf.Variable :<br>
tensorflow가 자체적으로 변경시키는 값<br> 
tensorflow가 학습하는 과정에서 스스로 변경시킴 == trainable variable

```python
# Out hypothesis WX+b
hypothesis = x_train * W + b
```

```python
# cost/Loss function
cost = tf.reduce_mean(tf.square(hypothesis - y_train))    # reduce_mean : 평균 내주는 것
```

```python
# Minimize -> GradientDescent
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)    # W와 b를 조절해서 cost를 minimize 해줌 / 그래프 상의 node임
```

# 2. Run/update graph and get results

```python
# Launch the graph in a session
sess = tf.Session()
```

```python
# Initializes global variables in the graph
# W, b라는 Variable을 사용 후 실행하기 전에 반드시 해줘야 함
sess.run(tf.global_variables_initializer())
```

```python
# Fit the line
for step in range(2001):
    sess.run(train)
    
    if step % 20 == 0:
        print(step, sess.run(cost), sess.run(W), sess.run(b))
```

### placeholder 사용하는 것으로 바꿔보기

```python
W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
```

```python
X = tf.placeholder(tf.float32, shape=[None])    # shape가 1차원 array이면서 몇개짜리든 가능하다는 말
Y = tf.placeholder(tf.float32, shape=[None])
```

```python
hypothesis = X * W + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))    # reduce_mean : 평균 내주는 것
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
# Fit the line
for step in range(2001):
    cost_val, W_val, b_val, _ = sess.run([cost, W, b, train], 
                                         feed_dict = {X: [1, 2, 3, 4, 5], Y: [2.1, 3.1, 4.1, 5.1, 6.1]})
    
    if step % 20 == 0:
        print(step, cost_val, W_val, b_val)
```

```python
# Testing our model
print(sess.run(hypothesis, feed_dict = {X: [5]}))
print(sess.run(hypothesis, feed_dict = {X: [2.5]}))
print(sess.run(hypothesis, feed_dict = {X: [1.5, 3.5]}))
```
