# Hypothesis using matrix

### matrix 사용 전

```python
import tensorflow as tf
```

```python
x1_data = [73., 93., 89., 96., 73.]
x2_data = [80., 88., 91., 98., 66.]
x3_data = [75., 93., 90., 100., 70.]

y_data = [152., 185., 180., 196., 142.]
```

```python
x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)

Y = tf.placeholder(tf.float32)
```

```python
w1 = tf.Variable(tf.random_normal([1]), name='weight1')
w2 = tf.Variable(tf.random_normal([1]), name='weight2')
w3 = tf.Variable(tf.random_normal([1]), name='weight3')

b = tf.Variable(tf.random_normal([1]), name='bias')
```

```python
hypothesis = x1 * w1 + x2 * w2 + x3 * w3 + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 1e-5)
```

```python
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001):
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train],
                                  feed_dict = {x1: x1_data, x2: x2_data, x3: x3_data, Y: y_data})
    
    if step % 10 == 0:
        print(step, "Cost: ", cost_val, "Prediction: ", hy_val)
```

# 파일 크기가 너무 클 때 -> Queue Runners

```python
filename_queue = tf.train.string_input_producer(
    ['data/data-01-test-score.csv'],  # 파일 이름들을 리스트화
    shuffle=False, 
    name='filename_queue'
)
```

```python
reader = tf.TextLineReader()
key, value = reader.read(filename_queue)
```

```python
record_defaults = [[0.], [0.], [0.], [0.]]
xy = tf.decode_csv(value, record_defaults=record_defaults)   # record_defaults : 읽어올 때 각 필드의 값이 어떤 형태인지
```

```python
train_x_batch, train_y_batch = tf.train.batch([xy[0:-1], xy[-1:]], batch_size=10)
```

```python
X = tf.placeholder(tf.float32, shape=[None, 3])
Y = tf.placeholder(tf.float32, shape=[None, 1])

W = tf.Variable(tf.random_normal([3, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

# Hypothesis
hypothesis = tf.matmul(X, W) + b

# Simplified cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))

# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)

# Launch the graph in a session.
sess = tf.Session()
# Initializes global variables in the graph.
sess.run(tf.global_variables_initializer())
```

```python
# Start populating the filename queue.
coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(sess=sess, coord=coord)
```

```python
for step in range(2001):
    x_batch, y_batch = sess.run([train_x_batch, train_y_batch])
    
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train],
                                  feed_dict = {X: x_batch, Y: y_batch})
    
    if step % 10 == 0:
        print(step, "Cost: ", cost_val, "Prediction: ", hy_val)
```

```python
coord.request_stop()
coord.join(threads)
```
